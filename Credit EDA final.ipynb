```python
#Explore datasets - application_data.csv - outliers
def outlier_graph(df, col, vartype):
    fig, ax=plt.subplots(nrows =1,ncols=2,figsize=(20,8))
    ax[0].set_title("Distribution Plot")
    sns.distplot(df_ad[col],ax=ax[0])
    ax[1].set_title("Box Plot")
    sns.boxplot(data =df_ad, x=col,ax=ax[1],orient='v')
```

```python
#Explore datasets - application_data.csv - outliers
col_graph = df_ad[['AMT_INCOME_TOTAL', 'AMT_CREDIT']]
for item in col_graph:
    outlier_graph(df=df_ad,col=item,vartype=1)
```

```python
#Explore datasets - application_data.csv - create bins for AMT_INCOME_TOTAL
bins = [0, 50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, 1000000]
slot = ['0-50K', '50K-10K', '100K-150K', '150K-200K', '200K-250K','250K-300K','300K-350K','350K-400K','400K-450K','450K-500K','500K and above']
df_ad['INC_RANGE'] = pd.cut(df_ad['AMT_INCOME_TOTAL'], bins, labels=slot)
```

```python
#Explore datasets - application_data.csv - create bins for AMT_CREDIT
bins = [0, 100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000, 1100000, 1200000, 1300000, 1400000, 1500000, 10000000]
slot = ['0-0.1m','0.1m-0.2m','0.2m-0.3m','0.3m-0.4m','0.4m-0.5m','0.5m-0.6m','0.6m-0.7m','0.7m-0.8m','0.8m-0.9m', '0.9m-1.0m', '1.0m-1.1m', '1.1m-1.2m', '1.2m-1.3m', '1.3m-1.4m', '1.4m-1.5m', '1.5m and above']
df_ad['CREDIT_RANGE'] = pd.cut(df_ad['AMT_CREDIT'], bins, labels=slot)
```

```python
##Univariate analysis - application_data.csv - create function to create graphs in one go for binned columns
col_graph = df_ad[['INC_RANGE', 'CREDIT_RANGE']]
for item in col_graph:
    univar_graph(df=df_ad,col=item,vartype=1)
```

```python
##Bivariate analysis - application_data.csv - create function to create graphs
def bivar_graph(df, ind, col, val, vartype):
    chart = pd.pivot_table(data=df_ad, index=ind, columns=col, values='TARGET')
    plt.figure(figsize = (9,7))
    chart1 = sns.heatmap(round(chart*100,1), annot=True, cmap='YlOrRd')
    plt.yticks(rotation=0)
    bottom, top = chart1.get_ylim()
    chart1.set_ylim(bottom + 0.5, top -0.5)
    plt.ylabel(ind)
    plt.xlabel(col)
```

```python
##Multivariate analysis - application_data.csv - show graph
bivar_graph(df=df_ad,ind='NAME_CONTRACT_TYPE', col='CODE_GENDER', val='TARGET', vartype=1)
```
